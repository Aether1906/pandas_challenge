```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
numUniqueItems = len(purchase_data["Item ID"].unique())

averagePrice = purchase_data["Price"].mean()
countPrice = purchase_data["Price"].count()
totalPrice = purchase_data["Price"].sum()

summaryTableDF = pd.DataFrame({"Number of Unique Items":[numUniqueItems], "Avg Item Price":averagePrice, "Number of Purchases": countPrice, "Total Revenue": totalPrice})
summaryTableDF["Avg Item Price"] = summaryTableDF["Avg Item Price"].map("${:,.2f}".format)
summaryTableDF["Total Revenue"] = summaryTableDF["Total Revenue"].map("${:,.2f}".format)

summaryTableDF
```

```python
countGenderDF = (purchase_data["Gender"].value_counts())
countGenderPer=(purchase_data["Gender"].value_counts()/countPrice) *100
genderFinal = pd.DataFrame({"Total Count":countGenderDF,"Percent of Players":countGenderPer})


genderFinal["Percent of Players"] = genderFinal["Percent of Players"].map("{:,.2f}%".format)
genderFinal
```


## Purchasing Analysis (Gender)

```python
genderPurchaseCount = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
#genderPurchaseCount
genderAvgPrice = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
#genderAvgPrice
genderTotalValue = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
#genderTotalValue
genderTotalPPP = genderTotalValue / genderFinal["Total Count"]

genderPurchasingDF = pd.DataFrame ({"Purchase Count": genderPurchaseCount, "Average Purchase Price":genderAvgPrice,
                    "Total Purchase Value":genderTotalValue, "Avg Total Purchase per Person": genderTotalPPP })

genderPurchasingDF["Average Purchase Price"] = genderPurchasingDF["Average Purchase Price"].map("${:,.2f}".format)
genderPurchasingDF["Total Purchase Value"] = genderPurchasingDF["Total Purchase Value"].map("${:,.2f}".format)
genderPurchasingDF["Avg Total Purchase per Person"] = genderPurchasingDF["Avg Total Purchase per Person"].map("${:,.2f}".format)

genderPurchasingDF
                                                                                                        
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Created a variable to lock in my 'SN' and 'Age'
ageCountDF = purchase_data.loc[:, ["SN", "Age"]].drop_duplicates()
#Created bins and labels for my ages
ageBins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 100]
groupLabels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#Created a column called 'Age Range' to display the respective age group each 'SN' fell within
ageCountDF["Age Range"] = pd.cut(ageCountDF["Age"], ageBins, labels = groupLabels)
#ageCountDF
```

```python
#Displayed totals of all ages that fell within each group
ageCountTotals = ageCountDF["Age Range"].value_counts()
ageCountTotals

#Formula to create Percentages of all age groups
agePercent = ageCountTotals / PlayerNum *100

#Formatting and labeling of each column
ageDemoDF = pd.DataFrame({"Total Counts":ageCountTotals, "Percentage of Players": agePercent})
ageDemoDF["Percentage of Players"] = ageDemoDF["Percentage of Players"].map("{:,.2f}%.".format)
ageDemoDF

#Organize Data
ageDemoDF.sort_index()
```

## Top Spenders

```python
topCustomer = purchase_data.groupby(["SN"]).count()["Price"]
topCustomerAvg = purchase_data.groupby(["SN"]).mean()["Price"]
topCustomerTotal = purchase_data.groupby(["SN"]).sum()["Price"]
customerSpendingDF = pd.DataFrame({"Purchase Count": topCustomer, "Average Purchase Price":topCustomerAvg, "Total Purchase Value": topCustomerTotal})

customerSpendingDF = customerSpendingDF.sort_values("Total Purchase Value", ascending=False)

customerSpendingDF["Average Purchase Price"] = customerSpendingDF["Average Purchase Price"].map("${:,.2f}".format)
customerSpendingDF["Total Purchase Value"] = customerSpendingDF["Total Purchase Value"].map("${:,.2f}".format)


customerSpendingDF.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
itemData = purchase_data.loc[:, ["Item ID", "Item Name", "Price"]]

itemCount = itemData.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
itemPrice = itemData.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")
itemTotal = itemData.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

itemDF = pd.DataFrame({"Purchase Count": itemCount, "Item Price": itemPrice, "Total Purchase Value": itemTotal})

mostPopularItemDF =itemDF.sort_values("Purchase Count", ascending=False)

mostPopularItemDF["Item Price"] = mostPopularItemDF["Item Price"].map("${:,.2f}".format)
mostPopularItemDF["Total Purchase Value"] = mostPopularItemDF["Total Purchase Value"].map("${:,.2f}".format)
mostPopularItemDF.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
mostProfitableDF = itemDF.sort_values("Total Purchase Value", ascending=False)

mostProfitableDF["Item Price"] = mostProfitableDF["Item Price"].map("${:,.2f}".format)
mostProfitableDF["Total Purchase Value"] = mostProfitableDF["Total Purchase Value"].map("${:,.2f}".format)
mostProfitableDF.head()
```
